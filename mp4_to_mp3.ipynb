```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_audio
import os

def convert_mp4_to_mp3(input_file, output_file):
    try:
        # Extract audio from the MP4 file and save it as MP3
        ffmpeg_extract_audio(input_file, output_file)
        print(f"Conversion successful. Saved as {output_file}")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    input_file = "/content/drive/MyDrive/picss/Screen.mp4"

    # Specify the output path in your Google Drive folder
    output_path = "/content/drive/MyDrive/picss/out.mp3"

    convert_mp4_to_mp3(input_file, output_path)
```
