# Data Visualization

# Line plot

```python
df = var.groupby("sex") ["bill_length_mm"].mean()
plt.plot(df.index, df.values, marker="o", color="red")
plt.title("Tranding the sex")
plt.xlabel("male")
plt.ylabel("female")
plt.grid(True, alpha=0.3)
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(5,3))


plt.scatter(var["bill_length_mm"], var["bill_depth_mm"], alpha=0.6, color="blue")
plt.title("Scatter Plot of bill_length_mm vs bill_depth_mm")
plt.xlabel("bill_length_mm")
plt.ylabel("bill_depth_mm")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()        
```

# Box plot

```python
sns.boxplot(x = "sex", y = "flipper_length_mm", data = var) #, order["","",""]
plt.title("sex on the different flipper_length_mm	")
plt.xlabel("male")
plt.ylabel("female")
plt.show()           
```

# Countplot

```python
sns.countplot(x = "sex", data = var)
plt.show()
```

# Violoinplot

```python
sns.violinplot(x = "sex", y = "flipper_length_mm", data = var)
plt.show()              
```

# Pie chart 

```python
# pie chart - sex of flipper_length_mm statua
sex_counts = var["bill_length_mm"].value_counts().head(15)


#figsize=(5,6) means the figure will be inches wide and 6 inches tall
plt.figure(figsize=(5,6))


plt.pie(sex_counts, labels=sex_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("sex status Distribution")
plt.show()
```

# Pair chart

```python
sns.pairplot(var, hue = "sex")
plt.show()
```
